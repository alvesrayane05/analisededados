### 2.1. Importando as bibliotecas necessárias e realizando a leitura dos arquivos <i>.csv</i>
<hr>
<p>O primeiro passo do projeto foi a realização dos downloads dos arquivos <i>.csv</i> com as informações dos discentes ingressantes na UFRN entre os anos de 2013 a 2018. Uma coisa que notamos durante esse download foi que o arquivo que continha as informações dos ingressantes de 2017 não estava disponível no Portal. Portanto, vale a pena salientar que esse estudo não abrange esse grupo de discentes.</p>
<p>Posteriormente, importamos as biliotecas necessárias para a leitura dos artivos e as análises dos dados que os compõem; bem como realizamos a leitura desses arquivos e armazenamos seus dados em uma variável para os ingressantes de cada ano.</p>

```python
#Importação das bibliotecas necessárias para a realização do notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Leituras dos CSV com os dados dos alunos do intervalo de 2013 a 2018, com exceção de 2017
dados_ingressantes_2013 = pd.read_csv("CSV/discentes-2013.csv", sep=';', usecols = ["nome_discente", "matricula", "ano_ingresso","status","sigla_nivel_ensino","nome_curso"])
dados_ingressantes_2014 = pd.read_csv("CSV/discentes-2014.csv", sep=';', usecols = ["nome_discente", "matricula", "ano_ingresso","status","sigla_nivel_ensino","nome_curso"])
dados_ingressantes_2015 = pd.read_csv("CSV/discentes-2015.csv", sep=';', usecols = ["nome_discente", "matricula", "ano_ingresso","status","sigla_nivel_ensino","nome_curso"])
dados_ingressantes_2016 = pd.read_csv("CSV/discentes-2016.csv", sep=';', usecols = ["nome_discente", "matricula", "ano_ingresso","status","sigla_nivel_ensino","nome_curso"])
dados_ingressantes_2018 = pd.read_csv("CSV/discentes-2018.csv", sep=';', usecols = ["nome_discente", "matricula", "ano_ingresso","status","sigla_nivel_ensino","nome_curso"])
```

### 2.2. Limpando do Dataset e filtando por nível de ensino e curso
<hr>
<p>Depois de realizada a leitura dos arquivos, removemos dos datasets os dados nulos e duplicados e realizamos a filtragem dos alunos com nível de ensino igual a Graduação (G) e que estivessem cursando Tecnologia da Informação.</p>
<p>Feito isso, criamos uma variável do tipo <i>list</i> nomeada de <i>concatenar</i> para, como o nome mesmo diz, concatenar todos os datasets em uma lista e convertê-los em um único dataset nomeado de <i>alunos_ingressantes_por_ano</i></p>

```python
# Limpando os dados nulos no dataset
dados_ingressantes_2013.dropna(inplace=True)
# Retirando os dados duplicados 
dados_ingressantes_2013 = dados_ingressantes_2013[~dados_ingressantes_2013.matricula.duplicated()]
# Filtrando turmas ingressantes da graduação
dados_ingressantes_2013 = dados_ingressantes_2013[dados_ingressantes_2013.sigla_nivel_ensino == 'G']
# Filtrando as turmas ingressantes o curso do BTI
dados_ingressantes_2013 = dados_ingressantes_2013[dados_ingressantes_2013.nome_curso == 'TECNOLOGIA DA INFORMAÇÃO']

# Limpando os dados nulos no dataset
dados_ingressantes_2014.dropna(inplace=True)
# Retirando os dados duplicados 
dados_ingressantes_2014 = dados_ingressantes_2014[~dados_ingressantes_2014.matricula.duplicated()]
# Filtrando turmas ingressantes da graduação
dados_ingressantes_2014 = dados_ingressantes_2014[dados_ingressantes_2014.sigla_nivel_ensino == 'G']
# Filtrando as turmas ingressantes o curso do BTI
dados_ingressantes_2014 = dados_ingressantes_2014[dados_ingressantes_2014.nome_curso == 'TECNOLOGIA DA INFORMAÇÃO']

# Limpando os dados nulos no dataset
dados_ingressantes_2015.dropna(inplace=True)
# Retirando os dados duplicados 
dados_ingressantes_2015 = dados_ingressantes_2015[~dados_ingressantes_2015.matricula.duplicated()]
# Filtrando turmas ingressantes da graduação
dados_ingressantes_2015 = dados_ingressantes_2015[dados_ingressantes_2015.sigla_nivel_ensino == 'G']
# Filtrando as turmas ingressantes o curso do BTI
dados_ingressantes_2015 = dados_ingressantes_2015[dados_ingressantes_2015.nome_curso == 'TECNOLOGIA DA INFORMAÇÃO']

# Limpando os dados nulos no dataset
dados_ingressantes_2016.dropna(inplace=True)
# Retirando os dados duplicados 
dados_ingressantes_2016 = dados_ingressantes_2016[~dados_ingressantes_2016.matricula.duplicated()]
# Filtrando turmas ingressantes da graduação
dados_ingressantes_2016 = dados_ingressantes_2016[dados_ingressantes_2016.sigla_nivel_ensino == 'G']
# Filtrando as turmas ingressantes o curso do BTI
dados_ingressantes_2016 = dados_ingressantes_2016[dados_ingressantes_2016.nome_curso == 'TECNOLOGIA DA INFORMAÇÃO']

# Limpando os dados nulos no dataset
dados_ingressantes_2018.dropna(inplace=True)
# Retirando os dados duplicados 
dados_ingressantes_2018 = dados_ingressantes_2018[~dados_ingressantes_2018.matricula.duplicated()]
# Filtrando turmas ingressantes da graduação
dados_ingressantes_2018 = dados_ingressantes_2018[dados_ingressantes_2018.sigla_nivel_ensino == 'G']
# Filtrando as turmas ingressantes o curso do BTI
dados_ingressantes_2018 = dados_ingressantes_2018[dados_ingressantes_2018.nome_curso == 'TECNOLOGIA DA INFORMAÇÃO']

# Dados para criar um novo DataFrame
concatenar = [dados_ingressantes_2013, dados_ingressantes_2014, dados_ingressantes_2015, dados_ingressantes_2016, dados_ingressantes_2018]
# Criação de um novo DataFrame com todos os dados reunidos
alunos_ingressantes_por_ano = pd.concat(concatenar)
```

### 2.3. Contando a quantidade de alunos por ano
<hr>
<p>Criamos, então, uma lista chamada de <i>quantidade_alunos_ano</i> com a finalidade de contar quantos descentes estavam cadastrados por ano no curso de BTI. Feito isso, transformamos esses dados em um dataset nomeado de <i>quantidade_alunos_ingressantes</i> e dividido em duas colunas: Ano e QuantidadeAluno.</p>

```python
# Variavel para saber a quantidade total de alunos ingressantes no BTI
quantidade_alunos_ano = []
# Quantidade de alunos ingressantes em 2013
quantidade_alunos_ano.append([2013, dados_ingressantes_2013.shape[0]])
# Quantidade de alunos ingressantes em 2014
quantidade_alunos_ano.append([2014, dados_ingressantes_2014.shape[0]])
# Quantidade de alunos ingressantes em 2015
quantidade_alunos_ano.append([2015, dados_ingressantes_2015.shape[0]])
# Quantidade de alunos ingressantes em 2016
quantidade_alunos_ano.append([2016, dados_ingressantes_2016.shape[0]])
# Quantidade de alunos ingressantes em 2018
quantidade_alunos_ano.append([2018, dados_ingressantes_2018.shape[0]])
# Quantidade de alunos ingressantes no BTI
alunos_ingressantes_por_ano.shape[0]

# Criação de um novo DataFrame com a quantidades de alunos ingressantes nos seus respectivos anos
quantidade_alunos_ingressantes = pd.DataFrame(quantidade_alunos_ano, columns=["Ano","QuantidadeAlunos"])
```

### 2.4. Separação e contagem dos alunos de acordo com seu status e ano de ingresso
<hr>
<p>Separamos, então, os alunos de acordo com seu status (<i>cancelado</i>, <i>trancado</i>, <i>concluído</i>, <i>formado</i> e <i>ativo</i>) e os dividimos em listas correspondentes. Depois, criamos um dicionário de cada ano (<i>de 2013 a 2018</i>) para armazenar a quantidade de alunos com cada status para quele ano de ingresso.</p>

```python
# Separação dos alunos de acordo com o seus status : CANCELADO, TRANCADO, CONCLUIDO, FORMADO ou ATIVO
alunos_cancelados = alunos_ingressantes_por_ano[alunos_ingressantes_por_ano["status"] == "CANCELADO"]
alunos_trancados = alunos_ingressantes_por_ano[alunos_ingressantes_por_ano["status"] == "TRANCADO"]
alunos_concluidos = alunos_ingressantes_por_ano[alunos_ingressantes_por_ano["status"] == "CONCLUÍDO"]
alunos_formados = alunos_ingressantes_por_ano[alunos_ingressantes_por_ano["status"] == "FORMADO"]
alunos_ativos = alunos_ingressantes_por_ano[alunos_ingressantes_por_ano["status"] == "ATIVO"]

# Criação de um dicionário de cada ano para armazenar a quantidade alunos que estão com seus status
# CANCELADO, TRANCADO, CONCLUIDO, FORMADO ou ATIVO no respectivo ano
situacao_aluno_2013 = {}
situacao_aluno_2013["ATIVO"] = alunos_ativos[alunos_ativos["ano_ingresso"] == 2013].shape[0]
situacao_aluno_2013["CANCELADO"] = alunos_cancelados[alunos_cancelados["ano_ingresso"] == 2013].shape[0]
situacao_aluno_2013["CONCLUIDO"] = alunos_concluidos[alunos_concluidos["ano_ingresso"] == 2013].shape[0]
situacao_aluno_2013["FORMADO"] = alunos_formados[alunos_formados["ano_ingresso"] == 2013].shape[0]
situacao_aluno_2013["TRANCADO"] = alunos_trancados[alunos_trancados["ano_ingresso"] == 2013].shape[0]

situacao_aluno_2014 = {}
situacao_aluno_2014["ATIVO"] = alunos_ativos[alunos_ativos["ano_ingresso"] == 2014].shape[0]
situacao_aluno_2014["CANCELADO"] = alunos_cancelados[alunos_cancelados["ano_ingresso"] == 2014].shape[0]
situacao_aluno_2014["CONCLUIDO"] = alunos_concluidos[alunos_concluidos["ano_ingresso"] == 2014].shape[0]
situacao_aluno_2014["FORMADO"] = alunos_formados[alunos_formados["ano_ingresso"] == 2014].shape[0]
situacao_aluno_2014["TRANCADO"] = alunos_trancados[alunos_trancados["ano_ingresso"] == 2014].shape[0]

situacao_aluno_2015 = {}
situacao_aluno_2015["ATIVO"] = alunos_ativos[alunos_ativos["ano_ingresso"] == 2015].shape[0]
situacao_aluno_2015["CANCELADO"] = alunos_cancelados[alunos_cancelados["ano_ingresso"] == 2015].shape[0]
situacao_aluno_2015["CONCLUIDO"] = alunos_concluidos[alunos_concluidos["ano_ingresso"] == 2015].shape[0]
situacao_aluno_2015["FORMADO"] = alunos_formados[alunos_formados["ano_ingresso"] == 2015].shape[0]
situacao_aluno_2015["TRANCADO"] = alunos_trancados[alunos_trancados["ano_ingresso"] == 2015].shape[0]

situacao_aluno_2016 = {}
situacao_aluno_2016["ATIVO"] = alunos_ativos[alunos_ativos["ano_ingresso"] == 2016].shape[0]
situacao_aluno_2016["CANCELADO"] = alunos_cancelados[alunos_cancelados["ano_ingresso"] == 2016].shape[0]
situacao_aluno_2016["CONCLUIDO"] = alunos_concluidos[alunos_concluidos["ano_ingresso"] == 2016].shape[0]
situacao_aluno_2016["FORMADO"] = alunos_formados[alunos_formados["ano_ingresso"] == 2016].shape[0]
situacao_aluno_2016["TRANCADO"] = alunos_trancados[alunos_trancados["ano_ingresso"] == 2016].shape[0]

situacao_aluno_2018 = {}
situacao_aluno_2018["ATIVO"] = alunos_ativos[alunos_ativos["ano_ingresso"] == 2018].shape[0]
situacao_aluno_2018["CANCELADO"] = alunos_cancelados[alunos_cancelados["ano_ingresso"] == 2018].shape[0]
situacao_aluno_2018["CONCLUIDO"] = alunos_concluidos[alunos_concluidos["ano_ingresso"] == 2018].shape[0]
situacao_aluno_2018["FORMADO"] = alunos_formados[alunos_formados["ano_ingresso"] == 2018].shape[0]
situacao_aluno_2018["TRANCADO"] = alunos_trancados[alunos_trancados["ano_ingresso"] == 2018].shape[0]

```

### 2.5. Contagem e cálculo das porcentagens do total de alunos por status
<hr>
<p>Por fim, realizamos a contagem discentes totais para cada status separados por seu ano de ingresso e os armazenamos em dicionários respectivos aos seus status. Além disso, com o valor total de alunos ingressantes e a quantidade total de discentes com seus respectivos status, calculamos a porcentagem de ocorrência de cada status e armazenamos esses resultados em listas para cada ano de ingresso.</p>

```python
# Criação dos dicionários para armazenar a quantidade de alunos a cada ano que estão no respectivo status do dicionário
alunos_ativos_totais = {}
alunos_cancelados_totais = {}
alunos_trancados_totais = {}
alunos_concluidos_totais = {}
alunos_formados_totais = {}

# Contagem e armazenamento dos alunos ativos nos anos de 2013,2014,2015,2016 e 2018
alunos_ativos_totais[2013] = dados_ingressantes_2013[dados_ingressantes_2013["status"] == "ATIVO"].shape[0]
alunos_ativos_totais[2014] = dados_ingressantes_2014[dados_ingressantes_2014["status"] == "ATIVO"].shape[0]
alunos_ativos_totais[2015] = dados_ingressantes_2015[dados_ingressantes_2015["status"] == "ATIVO"].shape[0]
alunos_ativos_totais[2016] = dados_ingressantes_2016[dados_ingressantes_2016["status"] == "ATIVO"].shape[0]
alunos_ativos_totais[2018] = dados_ingressantes_2018[dados_ingressantes_2018["status"] == "ATIVO"].shape[0]

# Contagem e armazenamento dos alunos cancelados nos anos de 2013,2014,2015,2016 e 2018
alunos_cancelados_totais[2013] = dados_ingressantes_2013[dados_ingressantes_2013["status"] == "CANCELADO"].shape[0]
alunos_cancelados_totais[2014] = dados_ingressantes_2014[dados_ingressantes_2014["status"] == "CANCELADO"].shape[0]
alunos_cancelados_totais[2015] = dados_ingressantes_2015[dados_ingressantes_2015["status"] == "CANCELADO"].shape[0]
alunos_cancelados_totais[2016] = dados_ingressantes_2016[dados_ingressantes_2016["status"] == "CANCELADO"].shape[0]
alunos_cancelados_totais[2018] = dados_ingressantes_2018[dados_ingressantes_2018["status"] == "CANCELADO"].shape[0]

# Contagem e armazenamento dos alunos concluídos nos anos de 2013,2014,2015,2016 e 2018
alunos_concluidos_totais[2013] = dados_ingressantes_2013[dados_ingressantes_2013["status"] == "CONCLUÍDO"].shape[0]
alunos_concluidos_totais[2014] = dados_ingressantes_2014[dados_ingressantes_2014["status"] == "CONCLUÍDO"].shape[0]
alunos_concluidos_totais[2015] = dados_ingressantes_2015[dados_ingressantes_2015["status"] == "CONCLUÍDO"].shape[0]
alunos_concluidos_totais[2016] = dados_ingressantes_2016[dados_ingressantes_2016["status"] == "CONCLUÍDO"].shape[0]
alunos_concluidos_totais[2018] = dados_ingressantes_2018[dados_ingressantes_2018["status"] == "CONCLUÍDO"].shape[0]

# Contagem e armazenamento dos alunos formados nos anos de 2013,2014,2015,2016 e 2018
alunos_formados_totais[2013] = dados_ingressantes_2013[dados_ingressantes_2013["status"] == "FORMADO"].shape[0]
alunos_formados_totais[2014] = dados_ingressantes_2014[dados_ingressantes_2014["status"] == "FORMADO"].shape[0]
alunos_formados_totais[2015] = dados_ingressantes_2015[dados_ingressantes_2015["status"] == "FORMADO"].shape[0]
alunos_formados_totais[2016] = dados_ingressantes_2016[dados_ingressantes_2016["status"] == "FORMADO"].shape[0]
alunos_formados_totais[2018] = dados_ingressantes_2018[dados_ingressantes_2018["status"] == "FORMADO"].shape[0]

# Contagem e armazenamento dos alunos trancados nos anos de 2013,2014,2015,2016 e 2018
alunos_trancados_totais[2013] = dados_ingressantes_2013[dados_ingressantes_2013["status"] == "TRANCADO"].shape[0]
alunos_trancados_totais[2014] = dados_ingressantes_2014[dados_ingressantes_2014["status"] == "TRANCADO"].shape[0]
alunos_trancados_totais[2015] = dados_ingressantes_2015[dados_ingressantes_2015["status"] == "TRANCADO"].shape[0]
alunos_trancados_totais[2016] = dados_ingressantes_2016[dados_ingressantes_2016["status"] == "TRANCADO"].shape[0]
alunos_trancados_totais[2018] = dados_ingressantes_2018[dados_ingressantes_2018["status"] == "TRANCADO"].shape[0]
```

```python
# Cria os arrays com os valores de status para cada ano de ingresso
pie_2013 = []
pie_2013.append(alunos_ativos_totais[2013]*100 / quantidade_alunos_ano[0][1])
pie_2013.append(alunos_cancelados_totais[2013]*100 / quantidade_alunos_ano[0][1])
pie_2013.append(alunos_concluidos_totais[2013]*100 / quantidade_alunos_ano[0][1])
pie_2013.append(alunos_formados_totais[2013]*100 / quantidade_alunos_ano[0][1])
pie_2013.append(alunos_trancados_totais[2013]*100 / quantidade_alunos_ano[0][1])

pie_2014 = []
pie_2014.append(alunos_ativos_totais[2014]*100 / quantidade_alunos_ano[1][1])
pie_2014.append(alunos_cancelados_totais[2014]*100 / quantidade_alunos_ano[1][1])
pie_2014.append(alunos_concluidos_totais[2014]*100 / quantidade_alunos_ano[1][1])
pie_2014.append(alunos_formados_totais[2014]*100 / quantidade_alunos_ano[1][1])
pie_2014.append(alunos_trancados_totais[2014]*100 / quantidade_alunos_ano[1][1])

pie_2015 = []
pie_2015.append(alunos_ativos_totais[2015]*100 / quantidade_alunos_ano[2][1])
pie_2015.append(alunos_cancelados_totais[2015]*100 / quantidade_alunos_ano[2][1])
pie_2015.append(alunos_concluidos_totais[2015]*100 / quantidade_alunos_ano[2][1])
pie_2015.append(alunos_formados_totais[2015]*100 / quantidade_alunos_ano[2][1])
pie_2015.append(alunos_trancados_totais[2015]*100 / quantidade_alunos_ano[2][1])

pie_2016 = []
pie_2016.append(alunos_ativos_totais[2016]*100 / quantidade_alunos_ano[3][1])
pie_2016.append(alunos_cancelados_totais[2016]*100 / quantidade_alunos_ano[3][1])
pie_2016.append(alunos_concluidos_totais[2016]*100 / quantidade_alunos_ano[3][1])
pie_2016.append(alunos_formados_totais[2016]*100 / quantidade_alunos_ano[3][1])
pie_2016.append(alunos_trancados_totais[2016]*100 / quantidade_alunos_ano[3][1])

pie_2018 = []
pie_2018.append(alunos_ativos_totais[2018]*100 / quantidade_alunos_ano[4][1])
pie_2018.append(alunos_cancelados_totais[2018]*100 / quantidade_alunos_ano[4][1])
pie_2018.append(alunos_concluidos_totais[2018]*100 / quantidade_alunos_ano[4][1])
pie_2018.append(alunos_formados_totais[2018]*100 / quantidade_alunos_ano[4][1])
pie_2018.append(alunos_trancados_totais[2018]*100 / quantidade_alunos_ano[4][1])
```

<p>O primeiro (<i><b>Gráfico 1</b></i>) é um gráfico de linhas que representa a relação da quantidade de discentes ingressantes em cada ano. No eixo das abscissas encontramos os anos analisados; e no eixo das ordenadas está disposto a quantidade de alunos ingressantes naquele ano. Já em 2016,  essa quantidade chegou à cerca de 350 ingressantes (20 alunos mais de um a outro), caracterizando o maior pico da quantidade de alunos nos anos registrados.</p>
<p>Contudo, apesar desse constante crescimento, em 2018 foram registrados pouco menos de 320 alunos no curso, o que indica uma queda de 30 novos alunos no curso de BTI em 2 anos.</p>
<p>É importante salientar que não possuímos os dados dos alunos ingressantes no ano de 2017. Apesar desse ano estar presente no gráfico, não possuímos essa informação e, portanto, deve-se desconsiderar a informação visual que o gráfico dispõe para esse ano.</p>
